```python
%matplotlib inline
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
fig = figure()
ax = Axes3D(fig)
x = np.arange(-5, 5, 1)
y = np.arange(-5, 5, 1)
x, y = np.meshgrid(x, y)
z = x**2 - y**2
ax.plot_wireframe(x, y, z)
#ax.scatter(0, 0, 0, 'x', 50, 'red')
show()
```
