```python
%matplotlib inline
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
fig = figure()
ax = Axes3D(fig)
x = np.arange(-5, 5, 1)
y = np.arange(-5, 5, 1)
x, y = np.meshgrid(x, y)
z = 2*x + 4*y
z1 = 2*x + 4*y + 50
ax.plot_wireframe(x, y, z)
ax.plot_wireframe(x, y, z1, color = "red")
#ax.scatter(0, 0, 0, 'x', 50, 'red')
show()

```
