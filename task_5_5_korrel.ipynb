```python
import numpy as np
import matplotlib.pyplot as plt
n = 100
r = 0.7
x = np.random.rand(n)
y = r*x + (1 - r)*np.random.rand(n)
plt.plot(x,y,'o')
plt.grid(True)
#chisl = np.sum((x-np.mean(x))*(y-np.mean(y)))
#znam = np.sqrt((np.sum((x-np.mean(x)**2)))*(np.sum((y-np.mean(y)**2))))
chisl = np.mean(x*y)-np.mean(x)*np.mean(y)
znam = np.sqrt(np.mean(x**2)-np.mean(x)**2)*np.sqrt(np.mean(y**2)-np.mean(y)**2)
print("Коэффициент корреляции = ", round(chisl/znam,2))
A = np.vstack([x, np.ones(len(x))]).T
a1, b1 = np.linalg.lstsq(A, y)[0]
plt.plot([0,1],[b1,a1+b1])
plt.show
```
