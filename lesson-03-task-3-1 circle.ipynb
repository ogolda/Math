```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
import numpy as np

x1= []
x2 = []
y1 = []
y2 = []

for i in range(1500):
    r = 1000

    x1.append(-i)
    x2.append(i)
    y = np.sqrt((r ** 2 - i ** 2))
    y1.append(y)
    y_ = -np.sqrt((r ** 2 - i ** 2))
    y2.append(y_)
       
plt.plot(x1, y1, color='r')
plt.plot(x1, y2, color='r')
plt.plot(x2, y2, color='r')
plt.plot(x2, y1, color='r')
plt.axis('scaled')
plt.show()
```
